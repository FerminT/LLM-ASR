```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
metrics = pd.read_csv('qwen2-1B_lora16_wavlm_layer10_instruction/logs_metrics/version_6/metrics.csv')
x = 'step'
y = ['training_loss']

for yi in y:
    plt.plot(metrics.loc[~metrics[yi].isna()][x],metrics.loc[~metrics[yi].isna()][yi], label=yi)
plt.legend()
```
